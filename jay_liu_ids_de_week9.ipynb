# Data Manipulation with Google Colab

This notebook demonstrates the Week 9 project for setting up and using Google Colab for data manipulation tasks. It includes loading data, performing descriptive analysis, and visualizing trends from the provided dataset.



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import logging

# Setup basic logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
````

````python
# Upload the dataset
uploaded = files.upload()

# Load the dataset into a pandas DataFrame
data = pd.read_csv('US_birth.csv')
logger.info("Dataset loaded successfully")

# Display the first few rows of the dataset
data.head()
````

````python
# Display dataset information
logger.info(f"Dataset Shape: {data.shape}")
logger.info(f"Columns: {data.columns}")

# Show data types and general info
data.info()

# Show basic statistics
data.describe()
````

````python
# Calculate mean and median of the 'births' column
mean_births = data['births'].mean()
median_births = data['births'].median()
logger.info(f"Mean Births: {mean_births}")
logger.info(f"Median Births: {median_births}")
````

````python
# Filter data for the year 2005
data_2005 = data[data['year'] == 2005]
logger.info(f"Filtered data for the year 2005:")
data_2005.head()
````

````python
# Plot birth trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='year', y='births', estimator='sum', ci=None)
plt.title('Total Births Over Years')
plt.xlabel('Year')
plt.ylabel('Total Births')
plt.show()
````

# Summary of Findings

- The mean number of births was found to be `mean_births`.
- The median number of births was `median_births`.
- The plot above illustrates the trend of total births over the years.
